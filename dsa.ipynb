### Time & Space Complexity

1. Time Complexity: Time complexity in computer science describes how the execution time of an algorithms grows as the input size increases. It essentially quantifies the number of operations an algorithm performs as a function of the input data size.
- Asymptotic Notation: 
    1. Best Case (Ω)
    2. Average Case (θ)
    3. Worst Case (O)




```python
x = int(input("x:"))
n = int(input("n:"))

for i in range(0,n):
    if i == x:
        break
    print(i)
    
"""
Here We can define a complexity of cases as we study. The best case is when the value of x is 0, and the worst case is when the value of x is n-1.
The average case is when the value of x is n/2. So we can say that the time complexity of this code is O(n) in the worst case, O(1) in the best case,
and O(n/2) in the average case.

"""
```

# Types of Time Complexity:



### 1. Linear Time (O(n))
👉 Algorithm ka running time input size (n) ke proportional badhta hai.

in simple word: If input become 2x then time must be 2x.

```python
for i in range(n): # O(n)
    print(i, end = " ")
```

```python
x = int(input("x:")) # O(n)

for i in range(x):
    for j in range(10):
        print(i,j, end=" ")
```

```python
# o(n+x)

for i in range(x):
    print(x, end=" ")

for i in range(n):
    print(n, end = " ")
    
# Here time complexity is sum of both loops O(n) + O(n) = O(2n) = O(n)
```

### 2. Constant Time (O(1))
👉 Algorithm ka running time input size (n) par depend nahi karta.

In simple word: Whether the input is 1 or 1 million, the algorithm will take the same time.

```python
x = 30000000
y = 2000000000

for i in range(3):
    for j in range(4):
        print(x,y,i,j, end= " ")
```

### 3. Quadratic Time (O(n^2))
👉 Algorithm ka running time input size (n) ke square ke proportional badhta hai.

In simple word: If input will 2x then the running time would be 4x.

```python
n = 4
m = 5

for i in range(n**2): # O(n^2)
    print(i, end = " ")
    
for i in range(n): # O(n^m)
    for j in range(m):
        print(i,j, end = " " )

for i in range(n): # O(n^2)
    for j in range(n):
        print(i,j, end = " " )
```

### 4. Logarithmic Time O(log n)

👉 Algorithm har step me problem size ko half (divide) kar deta hai.

In simple word: if input is 2x then the time increase less but not 4x 


# Array 

```python
import numpy as np

arr = input("Enter the array:").split()
arr = np.array(arr, dtype=int)

count = arr[0]

for i in range(len(arr)):
    if count > arr[i]:
        count = arr[i]

print("Array is:", arr)
print("Lowest Value of the array is:", count)
```

```python
# Sort the array and print the sorted array and the lowest value of the array

def sorted_arr(arr):
    n = len(arr)
    
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

ar = [5, 2, 9, 1, 5, 6]
print(sorted_arr(ar))

    
```
